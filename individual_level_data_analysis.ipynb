# The class, the object, the data and the methods

## Defining the class

```python
class IndividualLevelDataAnalysis:
    def loading_the_index_frame(self):
        self.final_frame = pd.read_csv('./individual_level_processed_data/final_frame.csv')
        
    def calculating_correlations(self):
        self.freedom_socialism = round(self.final_frame.groupby('country')[['freedom_index','socialism_index']].corr().unstack().iloc[:,1], 2)
        self.freedom_family = round(self.final_frame.groupby('country')[['freedom_index','family_index']].corr().unstack().iloc[:,1], 2)
        self.freedom_ethnicity = round(self.final_frame.groupby('country')[['freedom_index','ethnicity_index']].corr().unstack().iloc[:,1], 2)
        self.freedom_family = round(self.final_frame.groupby('country')[['freedom_index','family_index']].corr().unstack().iloc[:,1], 2)
        self.freedom_ethnicity = round(self.final_frame.groupby('country')[['freedom_index','ethnicity_index']].corr().unstack().iloc[:,1], 2)
        self.freedom_gender = round(self.final_frame.groupby('country')[['freedom_index','gender_index']].corr().unstack().iloc[:,1], 2)
        self.freedom_nationalism = round(self.final_frame.groupby('country')[['freedom_index','nationalism_index']].corr().unstack().iloc[:,1], 2)
        self.freedom_religiosity = round(self.final_frame.groupby('country')[['freedom_index','religiosity_index']].corr().unstack().iloc[:,1], 2)
        self.socialism_family = round(self.final_frame.groupby('country')[['socialism_index','family_index']].corr().unstack().iloc[:,1], 2)
        self.socialism_ethnicity = round(self.final_frame.groupby('country')[['socialism_index','ethnicity_index']].corr().unstack().iloc[:,1], 2)
        self.socialism_gender = round(self.final_frame.groupby('country')[['socialism_index','gender_index']].corr().unstack().iloc[:,1], 2)
        self.socialism_nationalism = round(self.final_frame.groupby('country')[['socialism_index','nationalism_index']].corr().unstack().iloc[:,1], 2)
        self.socialism_religiosity = round(self.final_frame.groupby('country')[['socialism_index','religiosity_index']].corr().unstack().iloc[:,1], 2)
        self.family_ethnicity = round(self.final_frame.groupby('country')[['family_index','ethnicity_index']].corr().unstack().iloc[:,1], 2)
        self.family_gender = round(self.final_frame.groupby('country')[['family_index','gender_index']].corr().unstack().iloc[:,1], 2)
        self.family_nationalism = round(self.final_frame.groupby('country')[['family_index','nationalism_index']].corr().unstack().iloc[:,1], 2)
        self.family_religiosity = round(self.final_frame.groupby('country')[['family_index','religiosity_index']].corr().unstack().iloc[:,1], 2)
        self.ethnicity_gender = round(self.final_frame.groupby('country')[['ethnicity_index','gender_index']].corr().unstack().iloc[:,1], 2)
        self.ethnicity_nationalism = round(self.final_frame.groupby('country')[['ethnicity_index','nationalism_index']].corr().unstack().iloc[:,1], 2)
        self.ethnicity_religiosity = round(self.final_frame.groupby('country')[['ethnicity_index','religiosity_index']].corr().unstack().iloc[:,1], 2)
        self.gender_nationalism = round(self.final_frame.groupby('country')[['gender_index','nationalism_index']].corr().unstack().iloc[:,1], 2)
        self.gender_religiosity = round(self.final_frame.groupby('country')[['gender_index','religiosity_index']].corr().unstack().iloc[:,1], 2)
        self.nationalism_religiosity = round(self.final_frame.groupby('country')[['nationalism_index','religiosity_index']].corr().unstack().iloc[:,1], 2)
                
    def correlation_matrix(self):
        self.country_correlations=pd.DataFrame(self.freedom_socialism)
        self.country_correlations=self.country_correlations.merge(self.freedom_family,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.freedom_ethnicity,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.freedom_gender,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.freedom_nationalism,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.freedom_religiosity,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.socialism_family,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.socialism_ethnicity,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.socialism_gender,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.socialism_nationalism,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.socialism_religiosity,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.family_ethnicity,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.family_gender,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.family_nationalism,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.family_religiosity,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.ethnicity_gender,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.ethnicity_nationalism,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.ethnicity_religiosity,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.gender_nationalism,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.gender_religiosity,left_index=True, right_index=True)
        self.country_correlations=self.country_correlations.merge(self.nationalism_religiosity,left_index=True, right_index=True)
        
    def renaming_the_columns(self):
        self.country_correlations.columns = ['freedom_socialism', 'freedom_family', 'freedom_ethnicity', 'freedom_gender', 'freedom_nationalism', 'freedom_religiosity',
                                            'socialism_family', 'socialism_ethnicity', 'socialism_gender', 'socialism_nationalism', 'socialism_religiosity', 
                                            'family_ethnicity', 'family_gender', 'family_nationalism', 'family_religiosity', 
                                            'ethnicity_gender', 'ethnicity_nationalism', 'ethnicity_religiosity', 
                                            'gender_nationalism', 'gender_religiosity', 
                                            'nationalism_religiosity']
        
    def saving_correlations(self):
        self.country_correlations.to_csv('./country_level_processed_data/country_correlations.csv', sep=',', index=True, encoding='utf-8')
```
